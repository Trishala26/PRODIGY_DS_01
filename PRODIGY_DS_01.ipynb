## Visualizing Categorical variables using Barplot

```python
#Check & plot 10 countries with maximum population in the year 1960 
max_20_pop_1960 = df.sort_values(by='1960', ascending=False).head(20)
```

```python
plt.figure(figsize=(6,4))
plt.bar(max_20_pop_1960['Country Name'], max_20_pop_1960['1960'], color='blue')
plt.title('Top 20 Countries by Population in the Year 1960')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Population')
```

```python
#Check & plot 10 countries with minimum population in the year 2003 

min_20_pop_2003 = df.sort_values(by='2003', ascending=True).head(20)
```

```python
plt.figure(figsize=(6,3))
plt.bar(max_20_pop_1960['Country Name'], min_20_pop_2003['2003'], color='orange')
plt.title('Top 10 Countries with minimum Population in the Year 2003')
plt.xlabel('Country')
plt.xticks(rotation=90)
plt.ylabel('Population')
```

```python
male_population=df[df['Indicator Code']=='SP.POP.TOTL']
male_population_sort=male_population.sort_values(by='2022' , ascending=False)
male_population_top=male_population_sort.head(15)
male_population_top
```

```python
female_population=df[df['Indicator Code'] == 'SP.POP.TOTL']
female_population_sort=female_population.sort_values(by='2022' , ascending=False)
female_population_top=female_population_sort.head(15)
female_population_top
```

```python
merged_population=pd.merge(male_population,female_population,on='Country Name' , suffixes=('_male','_female'))
merged_population['Total Population']=merged_population['2022_male'] + merged_population['2022_female']
total_merged_population=merged_population.sort_values(by='Total Population',ascending=False)
merged_population_top=total_merged_population.head(15)
merged_population_top
```

```python
plt.figure(figsize=(15,5))
sns.set(style='darkgrid')
plt.title('Top 15 Countries with Male and Female Populations (2022)')
sns.barplot(x='Country Name' , y='2022_male' , data=merged_population_top , color='red' , label='Male population' )
sns.barplot(x='Country Name' , y='2022_female' , data=merged_population_top , bottom=merged_population_top['2022_male'] ,label='Female population' , color='orange')
plt.legend()
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45,ha='right')
plt.show()
```

# Histograms for two groups of columns representing different time periods

```python
columns_group1 = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980']

columns_group2 = ['1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000']
bins = 30
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))

axes[0].hist(columns_group1, bins=bins, color='green', edgecolor='black', alpha=0.7)
axes[0].set_title('Histograms for  columns_group1')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Frequency')

axes[1].hist(columns_group2, bins=bins, color='violet', edgecolor='black', alpha=0.7)
axes[1].set_title('Histograms for  columns_group2')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
